```python
import pixellib
from pixellib.instance import instance_segmentation
```

```python
segment_video = instance_segmentation()
segment_video.load_model("mask_rcnn_coco.h5")
```

```python
segment_video.process_video("Hello.mp4", show_bboxes = True, frames_per_second= 5, output_video_name="output_detect.mp4")
```
